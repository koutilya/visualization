```python
with open("workdata/lines_submit/all_lines.txt",encoding = 'utf-8' ,mode ='r') as fr:
    lines = fr.readlines()

lines_ = []
fw1 = open("workdata/lines_submit/all_uniquelines.txt",encoding = 'utf-8' ,mode ='w') 
fw2 = open("workdata/lines_submit/all_repetedlines.txt",encoding = 'utf-8' ,mode ='w') 
for line in lines:
    if(line.startswith("filename : ")) :
        line1 = line
        continue
    if(line.startswith("lineno : ")) :
        line2 = line
        continue
    if(line.startswith("line : ")) :
        line3 = line
        line4 = line[7:]
    
        line4 = re.match(r'[ 0-9\*]*(.*)',line4).group(1)
        line4 = re.sub(r'\(i\+\)|\(i-\)|\(\+i\)|\(-i\)|\(i\)',"",line4)
        line4 = re.sub(r'\(-?[ =0-9\'\"\*\!\?]+|\(|[ =0-9\-\'\"\*\!\?]*\)'," ",line4)
        line4 = re.sub(r'\"|\?|\.|\,|\*|\!|_\?|\'',"",line4)
        line4 = re.sub(r' \ +'," ",line4)
        if not line4 in lines_ :
            lines_.append(line4)
           
            fw1.writelines(line1)
            fw1.writelines(line2)
            fw1.writelines("line : "+line4+"\n" )
        
        else:
            
            fw2.writelines(line1)
            fw2.writelines(line2)
            fw2.writelines("line : "+line4 +"\n")
fw1.close()
fw2.close()
```
